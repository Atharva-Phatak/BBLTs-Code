```python
import pandas as pd
import os
import re
import seaborn as sns
import matplotlib.pyplot as plt
import json
from wordcloud import WordCloud, STOPWORDS
import nltk

sns.set_context("paper")
sns.set_style("darkgrid")
sns.set(font_scale=2.5)
```

```python
%load_ext nb_black
```

```python
df = pd.read_csv("new_clean_data.csv")
```

```python
df.shape
```

```python
counts = df.Year.value_counts().sort_index()
counts_df = pd.DataFrame.from_dict(
    {
        "Number of Publications": counts.values.tolist(),
        "years": [int(x) for x in counts.index],
    }
)
```

```python
plt.subplots(figsize=(18, 18))
# sns.lineplot(data = )
ax = sns.barplot(x="years", y="Number of Publications", data=counts_df)
plt.xticks(rotation=90)
ax2 = ax.twinx()
ax2.plot(counts_df["Number of Publications"])
```

```python
plt.savefig("papers.png", dpi=800, bbox_inches="tight")
```
